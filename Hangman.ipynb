```python
# HANGMAN
# Wordlist exists as 'wordlist'. Otherwise player 2 can choose a word
# Hangman ASCII
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']
# Importing random to select a word from the list
# Importing getpass to hide user input
import random
import getpass

playing = ''

start = input("Are you ready to play Hangman?! Y or YES")

if start.lower()[0] == 'y':
    playing = True
else:
    playing = False

# Set up the random word/or get a user to input it

choose_word = input("Would player 2 like to choose a word (A), or would you like a word to be chosen at random? (B)")
if choose_word.lower()[0] == 'a':
    # Hiding user input using getpass
    word = getpass.getpass("Player 2, please type the word. Player 1 don't look!")
elif choose_word.lower()[0] == 'b':
    word = random.choice(wordlist)

lives = 6
display = ['_']*(len(word))
guess_word = list(word)
print(display)
    
while playing:
    guess = input("Guess a letter!")
    
    for i in range(len(word)):
        letter = word[i]
        if guess.lower() == letter:
            display[i] = letter
            print("Good guess!")
            
    if guess not in word:
        lives -= 1
        print(f"Unlucky! That was incorrect. {lives} lives remaining")
            
    print(display)
    print(stages[lives])
    
    if "_" not in display:
        print("Congrats, you won!")
        break
            
    elif lives == 0:
        print("Game Over!")
        break
     
    
replay = input("Do you want to keep playing? Y or N")
if replay.lower()[0] == 'y':
    pass
elif replay.lower()[0] == 'n':
    playing = False
```
